```python
import os
 
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)
 
```

```python
Image.MAX_IMAGE_PIXELS = 182659400
```

```python
url = '/content/drive/My Drive/Colab Notebooks/'
image1 = Image.open(url+'합침.tif')
print('원본 크기: ',image1.size) 
width, height = image1.size
```

```python
shift = 8
cutsize = 256<<(shift-1) # 4096
for left in range(0,width,cutsize):
  for up in range(0,height,cutsize):
    #crop(left,up, right, down)
    save_at_dir = url+str(shift)+'/'+str(up//cutsize)
    createFolder(save_at_dir);
    save_at_this = save_at_dir+'/'+str(left//cutsize)+'.jpg'

    자르기 = left,up,min(width,left+cutsize),min(height,up+cutsize)
    print(left, up, 'save at:',save_at_this,'left,up, right, down', *자르기)
    croppedImage=image1.crop(자르기)

    w, h = croppedImage.size
    croppedImage = croppedImage.resize((w>>(shift-1),(h>>(shift-1))))
    croppedImage.save(save_at_this, quality=96)

```
